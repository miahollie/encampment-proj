```python
## import pandas

import pandas as pd
```

```python
## read dataset into notebook

pd.read_excel("raw-data/arrest-crowd-data.xlsx")
```

```python
## store in a df 

df = pd.read_excel("raw-data/arrest-crowd-data.xlsx")
```

```python
## make new column for crowd size averages for each school

df['AVG-CROWD'] = df[["CROWD-SIZE-1","CROWD-SIZE-2","CROWD-SIZE-3"]].mean(axis=1)
```

```python
## call df

df
```

```python
## make a new column that shows arrests as a percentage of the average crowd size 

df['PCT-ARREST'] = df.apply(lambda x: x["ARRESTS-DETAINMENTS"] / x["AVG-CROWD"], axis = 1)
df
```

```python
## total number of arrests, by university type 

df.groupby("TYPE")["ARRESTS-DETAINMENTS"].sum().to_frame()
```

```python
## On average, what percentage of the crowds at public universities were arrested? 
## Private? 


df.groupby("TYPE")["PCT-ARREST"].mean().to_frame()
```

```python
## Which universities saw the most arrests?

df.sort_values("ARRESTS-DETAINMENTS", ascending = False).head(10)
```

```python
 ## which schools arrested the highest percentage of protesters, based on the ratio of arrests to crowd size?

df.sort_values("PCT-ARREST", ascending = False).head(10)
```

```python
## which schools arrested the smallest percentage of crowd members? 

df.sort_values("PCT-ARREST", ascending = True).head(10)
```

```python
## which cities arrested the highest percentage of protesters, based on the ratio of arrests to crowd size?\

df.groupby("CITY")["PCT-ARREST"].mean().sort_values(ascending = False).head(10).to_frame()
```

```python
## which states arrested the highest percentage of protesters, based on the ratio of arrests to crowd size?

df.groupby("STATE")["PCT-ARREST"].mean().sort_values(ascending = False).head(10).to_frame()
```
