```python
## store in df
df = pd.read_csv("data-projects/2024/group-encampment-project/raw-data/encampment-data.csv")
```

```python
## df info
df.info()
```

```python
## total number of arrests by university type

df.groupby("PRIVATE OR PUBLIC")["TOTAL ARRESTS / DETAINMENTS"].sum().to_frame()
```

```python
## total number of arrests by university

df.sort_values("TOTAL ARRESTS / DETAINMENTS", ascending = False).head(10)
```

```python
## total number of arrests by city 

df.groupby("CITY")["TOTAL ARRESTS / DETAINMENTS"].sum().sort_values(ascending = False).head(10).to_frame()
```

```python
## total number of encampments by city

df.value_counts("CITY", ascending = False).to_frame()
```

```python
## total number of encampments by state

df.value_counts("STATE", ascending = False).to_frame()
```

```python
## total number of arrests by state 

df.groupby("STATE")["TOTAL ARRESTS / DETAINMENTS"].sum().sort_values(ascending = False).to_frame()
```
